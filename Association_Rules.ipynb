```
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```
data = pd.read_excel('AR-CatalogCrossSell.xls')
data
```

```
data.drop(columns = 'Customer Number', inplace = True)
data
```

```
fp_result = fpgrowth.fpgrowth(data, min_support=0.4, use_colnames=True)
fp_result
```

```
fp_result = fpgrowth.fpgrowth(data, min_support=0.3, use_colnames=True)
fp_result
```

```
fp_result = fpgrowth.fpgrowth(data, min_support=0.2, use_colnames=True)
fp_result
```

```
fp_result = fpgrowth.fpgrowth(data, min_support=0.1, use_colnames=True)
fp_result
```

```
fp_association_rules = association_rules(fp_result, metric="confidence", min_threshold=0.7)
fp_association_rules
```

```
fp_result.to_csv('fp_result.csv')
fp_association_rules.to_csv('fp_association_rules.csv')
```
